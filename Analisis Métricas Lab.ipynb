```r
library(RColorBrewer,ggplot2)
library(reshape2)
palette(brewer.pal(5,"Set1"))
```

```r
dat_18_1=read.csv2("pr_lib_18_1.csv");str(dat_18_1)
# dat_18_1=rbind(dat_18_1,read.csv2("pr_lib_17_2.csv")) #Agregar registros 2017-2

nm_test=data.frame(table(dat_18_1$material))
nm_test$is_tst=nm_test$Freq>0
no_test=c(
	"",
	"DSM-5",
	"Ninguno",
	"NORMAS APA",
	"SPSS")
nm_test$is_tst=!(nm_test$Var1%in%no_test)
#nm_test=edit(nm_test);View(nm_test)

nm_mater=data.frame(table(dat_18_1$asignatur))
nm_mater$is_mtr=nm_mater$Freq>0
no_asig=c("","MERCADEO","NEGOCIOS INTERNACIONALES","SIN DEFINIR")
nm_mater$is_mtr=!(nm_mater$Var1%in%no_asig)
#nm_mater=edit(nm_mater);View(nm_mater)

dat_18_1=dat_18_1[
	dat_18_1$asignatur%in%nm_mater$Var1[nm_mater$is_mtr]&
	dat_18_1$material%in%nm_test$Var1[nm_test$is_tst],]

dat_1=dat_18_1[dat_18_1$mot_solic=="Consulta",c("semstre","material","asignatur")]
#write.csv2(dat_1, file="dat_1.csv")
```

```r
#################################
############Gráficos de barras
#################################

#####Pruebas
uso_pruebas=table(dat_1$material)#uso_pruebas=data.frame(table(dat_1$material))
uso_pruebas=uso_pruebas[
	order(uso_pruebas, decreasing = T)
]
uso_pruebas=uso_pruebas[uso_pruebas>0]
data.frame(uso_pruebas)
par(mai=c(1,3,1,1))
barplot(col=2,
	uso_pruebas[1:20],
	horiz=TRUE ,
	las=1, 
	cex.names=0.8, 
	main="Frecuencia de Uso de Pruebas"
	)
par(mai=c(1,1,1,1))
# write.csv2(data.frame(uso_pruebas), "uso_pruebas_17-2_18-1.csv")
```

```r

#####Asignatura
uso_asignatur=table(dat_1$asignatur)
uso_asignatur=uso_asignatur[
	order(uso_asignatur, decreasing = T)
]
uso_asignatur=uso_asignatur[uso_asignatur>0]
data.frame(uso_asignatur)
par(mai=c(1,3,1,1))
barplot(col=2,
	uso_asignatur,
	horiz=TRUE ,
	las=1, 
	cex.names=0.8, 
	main="Frecuencia de Consulta por asignatura"
	)
par(mai=c(1,1,1,1))


```

```r

#####Semestre
data.frame(table(dat_1$semstre))
barplot(col=2,
	table(dat_1$semstre),
	main="Frecuencia de uso por Semestre",
	xlab="Semestre")


```

```r

#################################
############Gráficos Cruzados
#################################

#### Usos Semestre

uso_semestre_tbl=(table(
	dat_1$material,
	dat_1$semstre))

uso_semestre_tbl=uso_semestre_tbl[rowSums(uso_semestre_tbl)>0,]
uso_semestre_tbl=uso_semestre_tbl[
	order(rowSums(uso_semestre_tbl), decreasing = T),
	]

max(uso_semestre_tbl)
bl_colors_2=c("white",colorRampPalette(c('#deebf7', '#08519c'))(max(uso_semestre_tbl)))#;plot(1:length(bl_colors_2),rep(0, length(bl_colors_2)), col=bl_colors_2, pch=16, cex=10)

par(mai=c(3,1,1,.1))
image(
	uso_semestre_tbl, col=bl_colors_2,
	xaxt = 'n',	yaxt = 'n', main="Uso de Pruebas por Semestre"
)
axis(
	las=2,cex.axis=0.7,
	labels=colnames(uso_semestre_tbl),
	side = 2,
	at=seq(0,1,length.out = length(colnames(uso_semestre_tbl)))#+(1/length(colnames(uso_semestre_tbl))/2)
)
axis(
	las=2,cex.axis=0.6,
	labels=rownames(uso_semestre_tbl),
	side = 1,
	at=seq(0,1,length.out = length(rownames(uso_semestre_tbl)))#+(1/length(colnames(uso_semestre_tbl))/2)
)
#legend("topright",title="Uso",legend=0:max(uso_semestre_tbl),fill=bl_colors_2, horiz=F, cex=0.45)
par(mai=c(1,1,1,1))


```

```r

####### Usos asignatura

uso_asig_tbl=(table(
	dat_1$material,
	dat_1$asignatur))

uso_asig_tbl=uso_asig_tbl[
	rowSums(uso_asig_tbl)>0,
	colSums(uso_asig_tbl)>0]
uso_asig_tbl=uso_asig_tbl[
	order(rowSums(uso_asig_tbl), decreasing = T),
	order(colSums(uso_asig_tbl), decreasing = T)
	]


max(uso_asig_tbl)
bl_colors=c("white",colorRampPalette(c('#deebf7', '#08519c'))(max(uso_asig_tbl)))#;plot(1:length(bl_colors),rep(0, length(bl_colors)), col=bl_colors, pch=16, cex=10)


par(mai=c(3,2,1,.1))
image(
	uso_asig_tbl, col=bl_colors,
	xaxt = 'n',	yaxt = 'n', main="Uso de Pruebas por Asignatura"
)
axis(
	las=2,cex.axis=0.7,
	labels=colnames(uso_asig_tbl),
	side = 2,
	at=seq(0,1,length.out = length(colnames(uso_asig_tbl)))#+(1/length(colnames(uso_asig_tbl))/2)
)
axis(
	las=2,cex.axis=0.6,
	labels=rownames(uso_asig_tbl),
	side = 1,
	at=seq(0,1,length.out = length(rownames(uso_asig_tbl)))#+(1/length(colnames(uso_asig_tbl))/2)
)
#legend("topright",title="Uso",legend=0:max(uso_asig_tbl),fill=bl_colors, horiz=F, cex=0.45)
par(mai=c(1,1,1,1))

```
