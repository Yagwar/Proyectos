# Análisis Omega escala

Omega para los items 2, 6, 8, 12, 13 y 14, que conforman un factor y para los items 1, 3, 4, 5, 7, 9, 10 y 11, que conforman el otro. Tuve que eliminar el ítem 4, así que porfa necesito el omega para este factor con el ítem 4 y sin el ítem 4

* Fuente: http://personality-project.org/r/psych/HowTo/omega.pdf


```r
# install.packages("psych",  repos="https://cloud.r-project.org/")
# install.packages("GPArotation",  repos="https://cloud.r-project.org/")
library(psych)
```

```r
data=read.csv2("PANSI.csv", header=T)
str(data)
```

# Factor 1
items 2, 6, 8, 12, 13 y 14,

```r
factor_1=data[c("P2","P6","P8","P12","P13","P14")]
str(factor_1)
```

```r
omega(factor_1)
```

# Factor 2
## Con reactivo 4

```r
factor_2A=data[c("P1","P3","P4","P5","P7","P9","P10","P11")]
str(factor_2A)
```

```r
omega(factor_2A)
```

## Factor 2 sin reactivo 4

```r
factor_2B=data[c("P1","P3","P5","P7","P9","P10","P11")]
str(factor_2B)
```

```r
omega(factor_2B)
```

# Omega total base de datos

```r
omega(data[2:ncol(data)])
```

## Escala completa sin reactivo 4

```r
omega(data[c(2:4,6:ncol(data))])
```
