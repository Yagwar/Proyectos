# Relaciones principales

```r
data_main=read.csv2("main_dat_tesis.csv")
str(data_main)
```

```r
for (clmn_indx in 2:(length(colnames(data_main))-1)){
    for (clmn_indx_2 in (clmn_indx+1):(length(colnames(data_main)))){
        crr_tst=cor.test(data_main[,clmn_indx],data_main[,clmn_indx_2])
        print(paste("Correlación entre", colnames(data_main)[clmn_indx],"y",colnames(data_main)[clmn_indx_2],
                    "r(x,y)=",round(crr_tst$estimate,3),
                    "valor p=",round(crr_tst$p.value,3),
                    "significativo a 0.05=",crr_tst$p.value<=.05))
        print(crr_tst)
#         jpeg(paste(colnames(dat_num)[clmn_indx],"-",colnames(dat_num)[clmn_indx_2],".jpg", sep=""),width = 600, height = 600)
        plot(data_main[,clmn_indx], data_main[,clmn_indx_2], pch=16, xlab=colnames(data_main)[clmn_indx],ylab=colnames(data_main)[clmn_indx_2])
        title(main=paste("Correlación entre", colnames(data_main)[clmn_indx],"y",colnames(data_main)[clmn_indx_2]),
              sub=paste("r(x,y)=",round(crr_tst$estimate,3),"valor p=",round(crr_tst$p.value,3)))
        abline(lm(data_main[,clmn_indx_2]~ data_main[,clmn_indx]))
#         dev.off()
    }
}
```

# Relaciones entre variables categóricas

```r
factores=c("Entidad","Estrato","Genero","Orientación.sexual","Tener.pareja","Tiempo")
vars_deps=c("Edad","apz_Reflexivo","apz_sensitivo","apz_intuitivo","apz_visual","apz_verbal","apz_secuencial","apz_global","atacante_fis","atacante_psico","victima_fis","victima_psico")
```

```r
for (factor in factores){
    for (var_dep in vars_deps){
        print("************************************************************")
        print("************************************************************")
        print("ANÁLISIS")
        print(paste("Variable categórica:",factor))
        print(paste("Variable numérica:",var_dep))
        print("************************************************************")
        lvn_tst=leveneTest(unlist(data[var_dep]) ~ unlist(data[factor]))
        aov_test <- aov(unlist(data[var_dep]) ~ unlist(data[factor]))
        p_val_lev=lvn_tst["group",'Pr(>F)']
        p_val_aov=summary(aov_test)[[1]][[1,"Pr(>F)"]]
        if (p_val_lev>.05){
            print(paste("prueba de levene, valor-p =", round(p_val_lev,3), ". Es mayor a 0.05; se asume homogeneidad de varianzas. Es viable comparar las medias de los grupos"))
#             print(lvn_tst)
        }else{
            print(paste("prueba de levene, valor-p =", round(p_val_lev,3), ". Es menor a 0.05; no se asume homogeneidad de varianzas. No es viable comparar las medias de los grupos" ))
#             print(lvn_tst)
        }
        if (p_val_aov<.05){
            print(paste("ANOVA de una vía, valor-p =", round(p_val_aov,3), ". Es menor a 0.05; Se encuentran diferencias estadísticamente significativas entre los grupos" ))
            print(summary(aov_test))
            print(TukeyHSD(aov_test))
        }else{
            print(paste("ANOVA de una vía, valor-p =", round(p_val_aov,3), ". Es mayor a 0.05; No se encuentran diferencias estadísticamente significativas" ))
            print(summary(aov_test))
        }
    }
}
```
