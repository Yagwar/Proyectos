```r
summary(datos)
```

```r
pairs(datos, pch=16)
```

```r
print ("Puntuaciones Típias de Semejanzas")
datos$SEMEJANZAS
print ("Promedio: ")
mean(datos$SEMEJANZAS)
print ("Desviación Estándar: ")
sd(datos$SEMEJANZAS)

max(datos$SEMEJANZAS)-min(datos$SEMEJANZAS)

##

hist(datos$SEMEJANZAS, 
     main="Histograma de Semejanzas",
     col="gray75", 
     xlab="Puntuación escalar", 
     ylab="Frecuencias",
     freq=F)
lines(density(datos$SEMEJANZAS, bw=sd(datos$SEMEJANZAS)),lty=3)
lines(density(datos$SEMEJANZAS),lty=2)
abline(v=mean(datos$SEMEJANZAS),lty=4)
```

```r
hist(datos$VOCABULARIO, 
     main="Histograma de Vocabulario",
     col="gray75", 
     xlab="Puntuación escalar", 
     ylab="Frecuencias",
     freq=F)
lines(density(datos$VOCABULARIO, bw=sd(datos$VOCABULARIO)),lty=3)
lines(density(datos$VOCABULARIO),lty=2)
abline(v=mean(datos$VOCABULARIO),lty=4)
```

```r
hist(datos$INFORMACION, 
     main="Histograma de Información",
     col="gray75", 
     xlab="Puntuación escalar", 
     ylab="Frecuencias",
    freq=F)
lines(density(datos$INFORMACION, bw=sd(datos$INFORMACION)),lty=3)
lines(density(datos$INFORMACION,),lty=2)
abline(v=mean(datos$INFORMACION),lty=4)
```

```r
plot(x=c(1:3),y=datos[1,], type="o", ylim=c(0,20), xlab="Subescalas", ylab="Puntuaciones escalares", main="Gráfico de resultados")
abline(h=c(7,13), lty=2, col="gray20")
for (i in c(2:nrow(datos))){
    lines(x=c(1:3),y=datos[i,], type="o", col=i)
    }
```
