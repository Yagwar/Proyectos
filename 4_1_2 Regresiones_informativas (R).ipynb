# Top 10 importancia $OR$ Top 10 Correlación

```r
pred_name="discret_bienestar"

n_preds=10
preds_columns_import=row.names(corr_inf_dist)[order(corr_inf_dist[paste(pred_name,"_importance_mean",sep="")],decreasing=TRUE)[1:n_preds]] # top importancia
preds_columns_correl=row.names(corr_inf_dist)[order(corr_inf_dist[pred_name],decreasing=TRUE)[1:n_preds]] # Top correlación
preds_columns=unique(c(preds_columns_import,preds_columns_correl,pred_name))

preds_columns=gsub(" ", ".",preds_columns)
preds_columns=gsub(",", ".",preds_columns)
preds_columns=gsub("-", ".",preds_columns)

print(preds_columns)

predictions_df=datos_norm[,colnames(datos_norm)%in%preds_columns]
# colnames(predictions_df)
# predictions_df[pred_name]=1-predictions_df[pred_name]# Invertir variable criterio

model_1 <- glm(discret_bienestar ~.,family=binomial(link='logit'),data=predictions_df)
summary(model_1)
summary(model_1)$coefficients
exp(cbind(OR = coef(model_1), confint(model_1)))
```

```r
##Identificar variables con relación inversa y Recodifcar variables inversas
predictions_resign=predictions_df
predictions_resign[,model_1$coefficients<0]=1-predictions_resign[,model_1$coefficients<0]
## instanciar modelo_2 con variables recodificadas.
model_2 <- glm(discret_bienestar ~.,family=binomial(link='logit'),data=predictions_resign)
summary(model_2)
summary(model_2)$coefficients
exp(cbind(OR = coef(model_2), confint(model_2)))
```
