### Comparaciones Edad con los percentiles de las pruebas

```r
comparar=c("VMI_P","PV_P","CM_P","TVPS_DIS","TVPS_MEM_","TVPS_SPA","TVPS_CON","TVPS_SEG","TVPS_FGR","TVPS_CLO")
perc_to_edmen= function(vector_percentil){
    vec_cat=c()
    for (x_percentil in vector_percentil){
        if(is.na(x_percentil)){
            vec_cat=c(vec_cat,NA)
        }else{
            if (x_percentil<15.9){
                vec_cat=c(vec_cat,"Inferior")
            }
            if(x_percentil>=015.9&x_percentil<84.1){
                vec_cat=c(vec_cat,"Adecuada")
            }
            if(x_percentil>=84.1){
                vec_cat=c(vec_cat,"Superior")
            }
        }
    }
    return(factor(vec_cat, levels=c("Inferior","Adecuada","Superior"), ordered=TRUE))
}
```

```r
for (per in comparar){
    print(paste("****",per,"****"))
    ed_mn=perc_to_edmen(datos_gral[,per])
    tbl_mn=table(ed_mn)
    print(tbl_mn)
    barplot(tbl_mn, main=per)
}
```
