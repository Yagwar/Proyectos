### Comparaciones Edad con los percentiles de las pruebas

```r
comparar=c("VMI_P","PV_P","CM_P","TVPS_DIS","TVPS_MEM_","TVPS_SPA","TVPS_CON","TVPS_SEG","TVPS_FGR","TVPS_CLO")
perc_to_edmen= function(vector_percentil){
    vec_cat=c()
    for (x_percentil in vector_percentil){
        if(is.na(x_percentil)){
            vec_cat=c(vec_cat,NA)
        }else{
            if (x_percentil<15.9){
                vec_cat=c(vec_cat,"Inferior")
            }
            if(x_percentil>=015.9&x_percentil<84.1){
                vec_cat=c(vec_cat,"Adecuada")
            }
            if(x_percentil>=84.1){
                vec_cat=c(vec_cat,"Superior")
            }
        }
    }
    return(factor(vec_cat, levels=c("Inferior","Adecuada","Superior"), ordered=TRUE))
}
```

```r
for (per in comparar){
    print(paste("****",per,"****"))
    ed_mn=perc_to_edmen(datos_gral[,per])
    tbl_mn=table(ed_mn)
    print(tbl_mn)
    barplot(tbl_mn, main=per)
}
```

## Tamaño de letra

```r
tam_let_categ=function(tm_let){
    cat=NULL
    if(is.na(tm_let)){cat=NA}else{
        if(tm_let>5){cat="A"}
        if(tm_let>3.5&tm_let<=5){cat="B"}
        if(tm_let>2.5&tm_let<=3.5){cat="C"}
        if(tm_let==2.5){cat="D"}
        if(tm_let<2.5){cat="E"}
    }
    return(cat)    
}
```

```r
data_tam_let=data_tale[,2:13]
for (rw in 1:nrow(data_tam_let)){
    for (clmn in 1:ncol(data_tam_let)){
        data_tam_let[rw,clmn]=tam_let_categ(as.numeric(data_tam_let[rw,clmn]))
    }
}
```

```r
for (i in 1:ncol(data_tam_let)){
    print(var_tip_tale[colnames(data_tam_let)[i]==var_tip_tale$ID_VAR,"etiq_var"])
    dtbl=table(data_tam_let[,i])
    print(dtbl)
    barplot(dtbl,main=var_tip_tale[colnames(data_tam_let)[i]==var_tip_tale$ID_VAR,"etiq_var"])
}
```

```r
prom_tam_let=NULL
for (i in rowMeans(data_tale[2:4], na.rm=T)){
    prom_tam_let=c(prom_tam_let,tam_let_categ(i))
}
barplot(table(prom_tam_let), main="Promedio General Tamaño de letra")
```

```r
dif_let_categ=function(dif_let){
    cat=NULL
    if(is.na(dif_let)){cat=NA}else{
        if(dif_let>=2){cat=2}
        if(dif_let<2){cat=1}
        if(dif_let==0){cat=0}

    }
    return(cat)    
}
```

```r
data_dif_let=data_tale[,c(7,10,13)]
for (rw in 1:nrow(data_dif_let)){
    for (clmn in 1:ncol(data_dif_let)){
        data_dif_let[rw,clmn]=dif_let_categ(as.numeric(data_dif_let[rw,clmn]))
    }
}
```

```r
superpos_categ=function(superpos){
    cat=NULL
    if(is.na(superpos)){cat=NA}else{
        if(superpos>10){cat=2}
        if(superpos>=4&superpos<=10){cat=1}
        if(superpos<3){cat=0}

    }
    return(cat)    
}
```

```r
data_superpos=data_tale[,23:25]
for (rw in 1:nrow(data_superpos)){
    for (clmn in 1:ncol(data_superpos)){
        data_superpos[rw,clmn]=superpos_categ(as.numeric(data_superpos[rw,clmn]))
    }
}
```

```r
var_ortog=variables_tale_test[variables_tale_test$ID_VAR%in%colnames(data_tale[,26:46]),]
ortog_dat=data_tale[,26:46]
colnames(var_ortog)
colnames(ortog_dat)
var_ortog$ID_VAR
var_ortog$subescala
```

```r
sbscl=NULL
ocrrncs=NULL
for(subesc in unique(var_ortog$subescala)){
    sbscl=c(sbscl,subesc)
    ocurrs=rowSums(ortog_dat[,var_ortog$subescala==subesc])
    ocrrncs=c(ocrrncs,sum(ocurrs,na.rm=T))
    hist(ocurrs,
         col="gray75", 
         main=paste ("Ortografía",subesc, "3 subtest"),
         xlab=subesc)
}
```

```r
par(mai=c(1,2,.1,.1))
barplot(ocrrncs,names.arg=sbscl, horiz=T, las=2)
data.frame(Error=sbscl,Ocurrencias=ocrrncs,Porcentaje=ocrrncs/sum(ocrrncs))
```
